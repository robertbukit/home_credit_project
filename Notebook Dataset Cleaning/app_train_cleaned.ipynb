```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('application_train.csv')
pd.set_option('display.max_columns', None)
```

```python
# pilih kolom kategori
categorical = df.select_dtypes(include='object')

#pilih kolom numerik
numerical = df.select_dtypes(include='number')

rows = df.shape[0]
cols = df.shape[1]

print(f'Jumlah baris: {rows}')
print(f'Jumlah kolom: {cols}')

# Cetak ringkasan statistik data kategorikal
print('\nRingkasan Statistik Data Kategorikal:')
display(categorical.describe().T)

# Cetak ringkasan statistik data numerik
print('\nRingkasan Statistik Data Numerik:')
display(numerical.describe().T)
```

```python
dfc = df.copy()

duplicate_rows = dfc.duplicated()
print(f'Jumlah baris duplikat: {duplicate_rows.sum()}')

# buat fungsi untuk menampilkan dataframe missing value
def check_null(dfc):
  null_values = dfc.isnull().sum()

  nvc = pd.DataFrame(null_values[null_values > 0].sort_values(), columns=['Total Null Values'])
  nvc['Percentage'] = (nvc['Total Null Values'] / len(dfc)) * 100
  nvc['Data Type'] = [dfc[col].dtype for col in nvc.index]

  # urutkan berdasarkan total nilai null dan persentasenya
  nvc.sort_values(by=["Total Null Values", "Percentage"], ascending=False, inplace=True)

  return nvc

result = check_null(dfc)
pd.set_option('display.max_rows', None)
result
```

```python
# 1. drop kolom dengan missing value > 50%
missing_threshold = 0.5
cols_to_drop = dfc.columns[dfc.isnull().mean() > missing_threshold]
dfc.drop(columns=cols_to_drop, inplace=True)

# 2. Imputasi kolom numerik dengan missing value antara 10% - 50% menggunakan median
num_cols = dfc.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    missing_pct = dfc[col].isnull().mean()
    if 0.1 < missing_pct <= 0.5:
        dfc[col].fillna(dfc[col].median(), inplace=True)

# 3. Imputasi kolom kategorikal dengan missing value antara 10% - 50% menggunakan modus
cat_cols = dfc.select_dtypes(include='object').columns
for col in cat_cols:
    missing_pct = dfc[col].isnull().mean()
    if 0.1 < missing_pct <= 0.5:
        dfc[col].fillna(dfc[col].mode()[0], inplace=True)

# 4. imputasi kolom dengan missing value dibawah 10%
cols_to_impute_less_than_10 = result[result['Percentage'] <= 10].index
for col in cols_to_impute_less_than_10:
  if dfc[col].dtype == 'object':
    dfc[col].fillna(dfc[col].mode()[0], inplace=True)
  else:
    dfc[col].fillna(dfc[col].median(), inplace=True)

# 5. cek hasil akhir
result = check_null(dfc)
result.style.background_gradient(cmap='Blues')
```
