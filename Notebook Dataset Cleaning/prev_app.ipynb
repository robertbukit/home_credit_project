```python
pd.set_option('display.max_columns', None)

# Pilih kolom-kolom bertipe objek dan simpan dalam variabel 'categoric'
categoric = df.select_dtypes('object')

# Pilih kolom-kolom bertipe numerik dan simpan dalam variabel 'numeric'
numeric = df.select_dtypes('number')

# Tentukan jumlah baris dan kolom
rows = df.shape[0]
cols = df.shape[1]

# Cetak bentuk DataFrame
print(f'Jumlah Baris: {rows}\nJumlah Kolom: {cols}')

# Cetak ringkasan statistik data kategorikal
print('\nRingkasan Statistik Data Kategorikal:')
display(categoric.describe().T)

# Cetak ringkasan statistik data numerik
print('\nRingkasan Statistik Data Numerik:')
display(numeric.describe().T)
```

```python
dfe = df.copy()
# Memeriksa baris duplikat
duplicate_rows = dfe.duplicated()

# Menampilkan jumlah baris duplikat
print(f"Jumlah baris duplikat: {duplicate_rows.sum()}")

# Fungsi untuk menghitung dan menampilkan kolom dengan nilai missing
def check_nan(dfe):
    # Hitung nilai null
    null_values = dfe.isnull().sum()

    # Buat DataFrame untuk total nilai null dan persentase
    nvc = pd.DataFrame(null_values[null_values > 0].sort_values(), columns=['Total Null Values'])  # Hanya ambil kolom dengan nilai null
    nvc['Percentage'] = (nvc['Total Null Values'] / dfe.shape[0]) * 100
    nvc["Data Type"] = [dfe[col].dtype for col in nvc.index]  # Ambil tipe data dari kolom yang ada

    # Urutkan berdasarkan total nilai null dan persentase
    nvc.sort_values(by=["Total Null Values", "Percentage"], ascending=False, inplace=True)

    # Tampilkan hasil dengan gaya
    return nvc.style.background_gradient(cmap='Blues')

# Panggil fungsi dan simpan hasilnya
styled_result = check_nan(dfe)

# Untuk menampilkan hasil di Jupyter Notebook, gunakan:
styled_result
```

```python
# 1. Drop kolom dengan missing value > 99%
dfe.drop(columns=['RATE_INTEREST_PRIMARY', 'RATE_INTEREST_PRIVILEGED'], inplace=True)

# 2. Imputasi numerik
dfe['AMT_DOWN_PAYMENT'].fillna(0, inplace=True)
dfe['RATE_DOWN_PAYMENT'].fillna(0, inplace=True)
dfe['DAYS_FIRST_DRAWING'].fillna(dfe['DAYS_FIRST_DRAWING'].median(), inplace=True)
dfe['DAYS_FIRST_DUE'].fillna(dfe['DAYS_FIRST_DUE'].median(), inplace=True)
dfe['DAYS_LAST_DUE_1ST_VERSION'].fillna(dfe['DAYS_LAST_DUE_1ST_VERSION'].median(), inplace=True)
dfe['DAYS_LAST_DUE'].fillna(dfe['DAYS_LAST_DUE'].median(), inplace=True)
dfe['DAYS_TERMINATION'].fillna(dfe['DAYS_TERMINATION'].median(), inplace=True)
dfe['NFLAG_INSURED_ON_APPROVAL'].fillna(0, inplace=True)
dfe['AMT_GOODS_PRICE'].fillna(dfe['AMT_GOODS_PRICE'].median(), inplace=True)
dfe['AMT_ANNUITY'].fillna(dfe['AMT_ANNUITY'].median(), inplace=True)
dfe['CNT_PAYMENT'].fillna(dfe['CNT_PAYMENT'].median(), inplace=True)
dfe['AMT_CREDIT'].fillna(dfe['AMT_CREDIT'].median(), inplace=True)

# 3. Imputasi kategorikal
dfe['NAME_TYPE_SUITE'].fillna('Unknown', inplace=True)
dfe['PRODUCT_COMBINATION'].fillna(dfe['PRODUCT_COMBINATION'].mode()[0], inplace=True)

# 4. drop kolom yang hanya memiliki 1 data hilang
dfe.drop(index=dfe[dfe.isnull().sum(axis=1) == 1].index, inplace=True)
```

```python
styled_result = check_nan(dfe)

styled_result
```
